```python
import pandas as pd 
data = pd.read_csv("student_performance_updated_1000.csv")
data.head(10)
```

```python
print("Columns:", data.columns.tolist())
```

```python
data.info()
```

```python
# Remove the students id and name as they are identifier
data.drop(columns=['StudentID','Name'],inplace=True)
# commetted above so that to not throw error "columns not exit as  i have executed it once"
data.head(10)
```

### Checking missing data

```python
data.isnull().sum()
```

### Handling Missing Data

```python
# We have to impute the data 
data.fillna(data.mean(numeric_only = True),inplace=True)
#  categorical dat parental support
data[['ParentalSupport','Gender','Online Classes Taken']] = data[['ParentalSupport','Gender','Online Classes Taken']].fillna('Unknown')
data.head()
```

```python
data.isnull().sum()
```

```python
# still having in gender and ParentalSupport,Online Classes Taken as they are not integer
# convert boolean onti integer
data[['Online Classes Taken']] = data[['Online Classes Taken']].replace({True:1,False:0})
data.head()
```

### Data Types

```python
data.dtypes
```

```python
numeric_cols = data.select_dtypes(include='number').columns
data[numeric_cols]=data[numeric_cols].round().astype(int)
data.dtypes
```

```python
categorical = data.select_dtypes(include='object').columns
numerical = data.select_dtypes(exclude='object').columns

print("Categorical Features:", list(categorical))
print("Numerical Features:", list(numerical))
```

```python
import numpy as np
data.loc[data['Study Hours'] < 0, 'Study Hours'] = np.nan
data['Study Hours'] = data['Study Hours'].fillna(data['Study Hours'].median())
data['Attendance (%)'] = data['Attendance (%)'].clip(0, 100)
data.describe()
```
